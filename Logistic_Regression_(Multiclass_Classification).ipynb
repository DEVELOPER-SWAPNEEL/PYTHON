```python
#In this file we will be dealing with visual letter detection.
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
dir(digits)
```

```python
#In the dir(digits) we get data as below haing 64 1D array, then images plot the graph
#Also there is this 'target' which matches the images with the data and provide us a target element(describes the images, says what the reality is!!)
 digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
#for printing a series of n no.s here for instance(n=5)
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
#we will use data and target to train out model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target,test_size=0.2)
len(X_train)
```

```python
len(X_test)
```

```python
#hence we successfully divided the whole test set into two parts of train 80% approx and 20 % test
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#Through the below line of code we have printed the image present in the sequence of images in the lib of sklearn and to confirm which digit it is actually we will use digits.target[500]
plt.matshow(digits.images[500])
digits.target[500]
```

```python
 #now we will predict it using the model
 model.predict([digits.data[500]])
```

```python
#to get overall model's accuracy we use confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, model.predict(X_test))
cm
```

```python
#For better visualisation we will be using seaborn similar to matplotlib
import seaborn as sn
plt. figure(figsize = (10,7))
sn. heatmap(cm, annot=True)
plt.xlabel ('Predicted' )
plt.ylabel ( 'Truth')
```
