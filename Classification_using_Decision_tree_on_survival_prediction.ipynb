```python
import pandas as pd
import numpy as np

from Logistic_regression import X_train, X_test, y_train

df=pd.read_csv('/Users/hare/Downloads/titanic.csv')
df
```

```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis=1)
df.head()
```

```python
df['Survived'].value_counts()
#here 0 says you die and 1 says you lived.
```

```python
input= df.drop('Survived',axis=1)
target = df['Survived']
input
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_sex = LabelEncoder()
le_age = LabelEncoder()
le_Fare = LabelEncoder()

input['le_Pclass_n'] = le_Pclass.fit_transform(input['Pclass'])
input['le_sex_n'] = le_sex.fit_transform(input['Sex'])
input['le_age_n'] = le_age.fit_transform(input['Age'])
input['le_Fare_n'] = le_Fare.fit_transform(input['Fare'])
input
```

```python
input.drop(['Pclass','Sex','Age','Fare'],axis="columns",inplace=True)
input
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(input,target,test_size=0.2)
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
model.score(X_test,y_test)
```
