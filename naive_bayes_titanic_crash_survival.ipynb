```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from Classification_using_Decision_tree_on_survival_prediction import target, X_test, y_train

df=pd.read_csv('/Users/hare/Downloads/titanic-2.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
inputs=df.drop('Survived',axis=1)
inputs
```

```python
sex_dummies =pd.get_dummies(inputs[['Sex']])
sex_dummies
```

```python
inputs=pd.concat([inputs,sex_dummies],axis="columns")
inputs
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
#now importing and using naive bayes gaussian model for training and testing the datasets
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
model.score(X_test,y_test)
```
